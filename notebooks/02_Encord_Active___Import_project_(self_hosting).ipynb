# üóùÔ∏è Generate SSH key pairs to get connected to the Encord platform

üìú You will reqire a [public-private key pair](https://docs.encord.com/sdk-and-api/api/getting-started#creating-user-authorization-header) for authentication when using the [Encord SDK](https://python.docs.encord.com/) or [Encord API](https://docs.encord.com/sdk-and-api/api/getting-started). 
You'd need to register the public key with Encord and use your private key as credentials.

````python
from pathlib import Path

private_key_file = Path.home() / ".ssh" / "encord_key"
public_key_file = private_key_file.with_suffix(".pub")
if not private_key_file.is_file(): 
    print("Generating key file")
    !ssh-keygen -t ed25519 -f $private_key_file -N '' -q
print(f"public key: {public_key_file.read_text().strip()}")

!encord-active config set ssh_key_path $private_key_file
````

### Steps to register public SSH key to encord

1. Open [Encord](https://app.encord.com).
2. Go to **User => Settings => Public keys => + New key**.
3. Enter
   **Temporary Colab Key**
   as **title** and paste your **public key**
   in the **public key field**.
4. Now, you can import your projects.

# üíª Choose an üü£ Encord Project


**You can utilize the [Python SDK](https://python.docs.encord.com/index.html) to retrieve the list of available projects from Encord that you have access to and review their details to make a selection.**

````python
!encord-active print encord-projects

# Or by code:
# print(get_projects_json(private_key_file))
````

## ‚è¨ Importing Encord Project


Now you are setup to import your projects.


From the list, choose the `project-hash` (key-parameter) of the project you want to import and paste it in the command below.

> Replace `<ENTER YOUR PROJECT HASH HERE>` with your project-hash parameter.

````python
!encord-active import project --project-hash "<ENTER YOUR PROJECT HASH HERE>"
````

## üèÅ Start the app with your own project



Encord Active will now have downloaded your data and run the prebuilt metric functions on your data.
The data will be stored in a subdirectory of your current working directory with the name of the project.

You will be able to see the `[name-of-your-project]` from the prints of the import process.

You can open the app with the following code block. It should open a new tab in your browser. If not, you can click the link printed to open it.

````python
!encord-active visualize -t "[name-of-your-project]"
````
